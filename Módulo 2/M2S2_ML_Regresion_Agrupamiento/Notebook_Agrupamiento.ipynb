## **Agrupamiento**

Un modelo de agrupamiento, como su nombre lo indica, es usado para agrupar instancias con características similares.

Ahora veamos un problema de agrupamiento. Usaremos K-Medios como método de clusterización. K-Medios lo veremos con mayor detalle en el curso Cómo resolver problemas con aprendizaje no supervisado.

```
#Importar módulos necesarios. Ya lo hemos hecho en la celda de arriba, pero para acostumbrarnos....
import matplotlib.pyplot as plt

# Datos. make_blobs crea grupo de datos
from sklearn.datasets import make_blobs
n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)

# Grafiquemos los datos que generó make_blobs
plt.scatter(X[:, 0], X[:, 1])
```

Es fácil observar que hay 3 grupos distintos. ¿Lo descubrirá el modelo?

```
from sklearn.cluster import KMeans
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X)

# Imprimamos los datos, pero ahora en diferente color de acuerdo con el resultado del modelo
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```
